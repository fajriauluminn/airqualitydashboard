### Data Cleaning

Rangkuman hal-hal yang harus dibersihkan:

Pada data df_aotizhongxin, terdapat beberapa variabel yang terdapat missing values yaitu PM2.5, PM10, SO2, NO2, CO, O3, TEMP, PRES, DEWP, RAIN, wd, dan WSPM. Maka kita akan mengganti missing values pada variabel tersebut dengan modus yang didapat dari masing-masing variabel

```python
nan_variable = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'wd', 'WSPM']
for variable in nan_variable:
    mode = df_aotizhongxin[variable].mode()[0]
    df_aotizhongxin[variable].fillna(mode, inplace=True)

print(df_aotizhongxin.isnull().sum())
```

Karena tipe data yang salah pada variabel year, month, day, hour terdapat pada semua dataframe, maka kita akan menggabungkan dataframe tersebut. Setelah itu kita ubah tipe data variabel tersebut menjadi tipe datetime

#### Penggantian tipe data pada variabel year, month, day, hour

```python
#penggabungan dataframe
df_airquality = pd.concat([df_aotizhongxin, df_changping, df_dingling, df_dongsi, df_guanyuan, df_gucheng, df_huairou, df_nongzhanguan, df_shunyi, df_tiantan, df_wanliu, df_wanshouxigong], ignore_index=True)
df_airquality
```

```python
#Pembuatan variabel baru 'datetime' dan penggantian format agar sesuai
df_airquality['datetime'] = pd.to_datetime(
    df_airquality['year'].astype(str) + '-' + 
    df_airquality['month'].astype(str).str.zfill(2) + '-' + 
    df_airquality['day'].astype(str).str.zfill(2) + ' ' + 
    df_airquality['hour'].astype(str).str.zfill(2) + ':00'
)

# Verifikasi hasil
df_airquality
```

```python
#Penghapusan variabel hour day month year
drop_columns = ['hour', 'day', 'month', 'year']
df_airquality.drop(columns = drop_columns, inplace=True)
df_airquality
```

### Mengurutkan rata-rata kadar setiap parameter

```python
#Karbon Monoksida
average_carbon = df_airquality.groupby('station')['CO'].mean().reset_index()
sorted_avg_carbon = average_carbon.sort_values(by='CO', ascending=True)
sorted_avg_carbon
```

```python
#PM 10
average_pm10 = df_airquality.groupby('station')['PM10'].mean().reset_index()
sorted_avg_pm10 = average_pm10.sort_values(by='PM10', ascending=True)
sorted_avg_pm10
```
